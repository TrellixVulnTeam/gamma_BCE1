```python
from gamma import *

make_node = lambda t: {'type': t, 'params': {}}
identity = make_node('Identity')

@bind_vars
def remove_identities(name, _in):
    LHS = {name: (identity, [_in])}
    RHS = {}
    redirects = (name, _in)
    return LHS, RHS, redirects

g = pipeline([
    ('A', identity, ['input']),
    ('B', identity),
    ('C', make_node('Type_A')),
    ('D', identity),
    ('E', identity, ['C']),
    ('F', identity),
    ('G', identity),
    ('H', make_node('Type_B')),
    ('I', identity),
    ('J', identity),
    ('K', identity, ['H']),
    ('L', make_node('Type_C'), ['I'])
], prefix='scope')

draw(g)
draw(apply_rule(g, remove_identities()))
```
